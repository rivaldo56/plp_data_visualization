```python
# Task 1: Load and Explore the Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset from sklearn
iris = load_iris(as_frame=True)
df = iris.frame   # convert to pandas DataFrame

# Display first 5 rows
print("First 5 rows of the dataset:")
print(df.head())

# Check structure
print("\nDataset Info:")
print(df.info())

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Summary statistics
print("\nBasic Statistics:")
print(df.describe())

# Group by species and compute mean of numerical columns
species_means = df.groupby("target").mean()
print("\nMean values grouped by species:")
print(species_means)

# Replace target numbers with species names for readability
df['species'] = df['target'].map({i: name for i, name in enumerate(iris.target_names)})

print("\nInsights:")
print("- Iris-setosa generally has the smallest petal and sepal measurements.")
print("- Iris-virginica has the largest petal length and width.")
```

```python
# 1. Line chart - example: petal length trend (first 50 rows)
plt.figure(figsize=(8,5))
plt.plot(df.index[:50], df['petal length (cm)'][:50], marker='o', label='Petal Length')
plt.title("Line Chart of Petal Length (first 50 samples)")
plt.xlabel("Sample Index")
plt.ylabel("Petal Length (cm)")
plt.legend()
plt.show()

# 2. Bar chart - average petal length per species
plt.figure(figsize=(8,5))
sns.barplot(x="species", y="petal length (cm)", data=df, estimator="mean")
plt.title("Average Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Avg Petal Length (cm)")
plt.show()

# 3. Histogram - distribution of sepal length
plt.figure(figsize=(8,5))
plt.hist(df["sepal length (cm)"], bins=15, color="skyblue", edgecolor="black")
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()

# 4. Scatter plot - Sepal Length vs Petal Length
plt.figure(figsize=(8,5))
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", hue="species", data=df)
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.show()
```
